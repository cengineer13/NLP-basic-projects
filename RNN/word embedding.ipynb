```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
voc_size =10000

sent=[ 'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

## 1. One-hot Encoding

```python
#Convert to one-hot represantation
one_hot_repr = [one_hot(input_text=words, n = voc_size) for words in sent]
one_hot_repr
```

10000 talik vocabulary dictionarydagi tepadagi so'zlarning indexsi. Ya'ni "the", "glass" so'zlari 644, 5644 orinda turadi 10000 talik vektorda va 1 tonda 644 orindagi "the" sozi o'zi 1 va qolgan barcha so'zlar 0 bo'ladi. ya'ni (1,10000) vectorda faqat 644 orinda turgan index 1 qolgani 0 boladi

## 2. Word Embedding

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

Sentence har xil uzunlikda so'zlar bo'ladi. RNN uchun bir xil input bolishi kerak shu sababli pad_seqences orqali bir xil ravishda oldiga yoki orqasiga 0 berib keltirib olamiz

```python
sent_length = 8  
embedded_data = pad_sequences(sequences=one_hot_repr, maxlen=sent_length, padding='pre') #pre-oldidan, post - keyin 
embedded_data
```

```python
#Feature represantation 
fdim = 10 #har bir soz 10 talik fazoda qiymatga ega boladi. Word2Vec 
```

```python
model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=fdim,input_length=sent_length))
model.compile('adam', 'mse')
model.summary()
```

```python
y = model.predict(embedded_data)
y
```
